### In this demonstration, we show how the KBase iPython notebook can be used to explore seqeunce quality information.

We'll look at three example samples:
<p>
One exhibits good quality ("qc_good")
<p>
Another exhibits poor quality ("qc_poor")
<p>
A third demonstrates how artifacts like adaptor contamination can be detetected ("qc_ugly").
<p>
We start by initializing the notebook:

```python
init_ipy()
```

#### Then we download precomputed QC data for three metagenomic sequence samples:

```python
qc_set = get_collection(mgids=['mgm4440640.3','mgm4473069.3','mgm4473205.3'])
qc_good = QC(metagenome=qc_set.metagenomes['mgm4440640.3'])
qc_bad  = QC(metagenome=qc_set.metagenomes['mgm4473069.3'])
qc_ugly = QC(metagenome=qc_set.metagenomes['mgm4473205.3'])
```

#### We can get an intial assessment of the samples by looking at their DRISEE (PMID 22685393) errors: 

```python
print "good : ",qc_good.drisee.summary['data'][6]
print "bad  : ", qc_bad.drisee.summary['data'][6]
print "ugly : ", qc_ugly.drisee.summary['data'][6]
```

#### We can get a much more detailed picture of the DRISEE error by looking at the full DRISEE error profiles:

```python
print "good"
qc_good.drisee.plot()

print "bad"
qc_bad.drisee.plot()

print "ugly"
qc_ugly.drisee.plot()
```

#### We can also look at the nucletide profiles for these data sets:

```python
print "good"
qc_good.bp_histo.plot()

print "bad"
qc_bad.bp_histo.plot()

print "ugly"
qc_ugly.bp_histo.plot()
```

The DRISEE and nucleotide profiles lead to similar conclusions, the good sample exhibits
reasonable seqeunce quality, and a random distribution of bases \(25% A,T,C, & G\), as we 
would expect for most community data.  The low quality and "ugly" samples both contain a 
substantial amount of adaptor contamination, \(apparant as biased dsitributions of A,T,C,& G\)
in the nucleotide profiles.  These samples also exhibit poor DRISEE errors.

<h4>The kmer spectra reveal additional information about the samples:</h4>
    
Kmer spectra permit us to rapidly identify sequence diversity and coverage ranges for sequence datasets.  This knowledge can inform samplecomparison and exploitation via clustering and assembly.
    
The kmer.plot_abundance method gives us graphical summaries of kmer abundance and diversity, plotting the logarithm of the kmer abundance against the logarithm of kmer rank.

```python
print "good"
qc_good.kmer.plot_abundance()
```

The kmer abundance graph from good shows us that the sequences
represented are diverse.  None of the biological seqeunces have
coverages over 100x; most of the sequence is approx 10x, and
sequencing adapters occur at 1000x.

```python
print "ugly"
qc_ugly.kmer.plot_abundance()
qc_ugly.kmer.plot_ranked()
```

The abundance graph for ugly, by contrast, shows an adapter abundance
of almost 100,000x.  These artifactual sequences consume about a third
of the sequencing yield and have unknown consequences on downstream
annotation.
