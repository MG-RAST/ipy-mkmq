## Raw data boxplot

We use the following command to generate boxplots for the raw counts of all 120 samples:

The data exhibit non normal distributions, each of each which are unique.  This will complicate comparative analyses, making it difficult to
<p>
    determine if differences observed among samples are due to biologically relevent factors, or to bias introduced by experimental and/or
    <p>
        sequencing error.
<p>    
<p>
    To account for this problem, we'll use the normalized values.  We downloaded normlized values along with the raw count data. We can 
 <p>
     look at them the same way that we looked at the raw count data.  Normalized values for each sample were calculated as follows:
 <p>
     All values were log transformed: x_trans = log2(x + 1)
 <p>
 The average for transformed values in each sample was determine \(x_mean\) as well as the standard deviation \(x_sd\)    
 <p>
     For all values in a sample i=1..n, normalized (i.e. standaridized) values were calculated using the follwing formula:
 <p>
     x\(i\)_norm = \( \( x\(i\) - x_mean \)/ x_sd \)         
 <p>
     This procedure corrects for differences in scale and location among the samples, making errant observations 
 <p>
     due to experimetnal/seqeuencing error less likely.
 <p>
     After this procedure, the final nomralized values across all samples are scaled from 0 to 1.
<p>    
<p>
    We can observe the log transformed, standardized, and scaled values like this:

```python
%%R
options (width = 180)
print (my_data$norm [1:20,1:10])
```

Looking at these values, it's difficult to tell if normalization had the desired effect.  The easiest way to check is to look at boxplots of the normalized values:

### Boxplots of normalized data

The following command will allow us to look at the ditributions of the normalized data as boxplots:

```python
%R --width 1400 --height 600 -n boxplot (my_data, view = "norm", main = "normalized abundance data boxplots", las = 2, names = samples (my_data))
```

Normalization has removed a substantial amount of the bias in the abundance data for each sample.  Now, the distribution of count data for each sample 
<p>
is similar.  When we use the normalized values in our comparative analyses, we are much less likley to make erroneous conclusions based on differences 
<p>
in the data distributions driven by biologically un-interesting error (e.g. differences in sequencing depth between samples)

### Using Metadata to explore and analyze the data:

Now that we've processed the data to reduce the effect of experimental error, we can perform analyses to determine
if there are trends that exhibit correlation with respect to the three factors under consideration: 
<ul>
    <li> treatment: CC=continuous, P=unfertilized prarie, PF=fertilized prarie </li>
    <li> fraction:  WS=whole soil, LM=large macroaggregate, MM = medium macroaggregate, SM=small macroaggregate, Micro=micro aggregate </li>
    <li> date:      October or July of 2012 </li>
</ul>
<p>
    Below, R/matR is used to group data by these metadata, and we print a summary of these metadata for each sample:

```python
%%R
sample.data <- metadata (my_data) ["sample.name"]
sample.data <- strsplit (sample.data, "-", fixed = FALSE)
names (sample.data) <- substr (names (sample.data), 1, 12)
sample.data <- t (simplify2array (sample.data))
colnames (sample.data) <- c ("Treatment", "Fraction", "Date")
 
for (i in 1:dim(sample.data)[[1]]){
    substring_i <- substr (sample.data [i, "Treatment"], 1, 2)                         
    if ( (substring_i == "PF" | substring_i == "CC") ){
        sample.data[i,"Treatment"] <- substring_i
    }else{
          sample.data[i,"Treatment"] <- "P"
    }                                                                       
}
 
treatment.groups <- sample.data[,"Treatment"]
fraction.groups <- sample.data[,"Fraction"]
date.groups <- sample.data[,"Date"]
 
sample.data <- sample.data[ order(sample.data[,1]), ]
 
print(sample.data[1:10,])
```

### Statistical evaluation and subselection of the data.

First, we load a function to perform q-value (PMID: 12883005) based assessment of the false discovery rate below:

```python
%%R
source("/kb/deployment/services/analysis_book/notebook/lib/qvalue_function.r")
```

We perform a statistical test \(the Mann-Whitney test \(e.g. non-parametric T-test\)\) to find the taxa that exhibit the most
<p>
significa difference between the continuos corn (CC) and prarie treatments (unfertilized \(P\) and fertilized \(PF\) prarie\).
    <p>
        First, we take the treatment.groups from above and modify them, so that there are two groups \(CC and P + PF\) 

```python
%%R

for (i in 1:length(treatment.groups)){
    if ( treatment.groups[i] == "PF" ){
        treatment.groups[i] <- "P"
    }
} 

treatment.groups <- treatment.groups[dimnames(my_data$norm)[[2]]]
```

Now we use the simplified groupings \(CC vs P\) to identify the taxa that exhibit the most significant differences between these two groups:
<p> \(Results are sorted by p value\)

```python
%%R
groups (my_data) <- treatment.groups
 
sig.results <- sigtest (my_data, test = "Mann-Whitney")
q.results <- qvalue (sig.results$p.value, fdr.level = 0.001)
results <- data.frame (Annotation = rownames (my_data), stat = sig.results$stat, p.value = sig.results$p.value, q.value = q.results$qvalues, significant = q.results$significant)

results <- results[ order(results[,'p.value']), ]

print("Signficant = TRUE means that the indiacted taxonomic class is signficnatly differenciated between the CC and P/PF groups with FDR <= 0.001")
print("Results:")
print (results[1:10,])
```

We can see that a number of the taxa exhibit highly significant differences between the CC and P/PF groups.
<p>
   30 are signficant with and FDR <= 0.001 Now we'll subselect the singifcant data:
    <p>
        We can look at the data as a table like this:

```python
%R --width 1200 --height 600 -n heatmap (my_data, rows = results$significant, main = "normalized counts\nwith statistically driven row subselection", labCol = groups (my_data), cexCol = 1, col = hm.colors, colsep = NULL)
```

Even when we subselect for the taxa that exhibit the most significant differences between the CC and P/PF treatments, the trends that
<p>
    separate these groups are sutble -- furhter investigations (like those Prof. Hofmockel presented) enabled by these tools are necessary
    <p> to understad the taxonomic trends.
